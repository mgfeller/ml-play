9.3 Grid Search Deep Learning Model Parameters

```python
# Setup 
from keras.models import Sequential
from keras.layers import Dense

from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score

import numpy as np
import tensorflow as tf
print(tf.__version__)
seed = 7
np.random.seed(seed)
```

```python
# Load dataset
dataset = np.loadtxt("pima-indians-diabetes.csv", delimiter=",")
X = dataset[:,0:8]
Y = dataset[:,8]
```

```python
# Function to create model, required for KerasClassifier
def create_model(optimizer='rmsprop', init='glorot_uniform'):
    # Create model
    model = Sequential()
    model.add(Dense(12, input_dim=8, kernel_initializer=init, activation='relu'))
    model.add(Dense(8, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    # Compile model
    model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model
```

```python
# Create model (create model signature: create_model(optimizer='rmsprop', init='glorot_uniform'))
model = KerasClassifier(build_fn=create_model, verbose=0)
```

```python
# Grid search optimizers, initializers, number of epochs, batch size
from sklearn.model_selection import GridSearchCV

optimizers = ['rmsprop', 'adam']
inits = ['glorot_uniform', 'normal', 'uniform']
epochs = [50, 100, 150]
batches = [5, 10, 20]
param_grid = dict(optimizer=optimizers, epochs=epochs, batch_size=batches, init=inits)
grid = GridSearchCV(estimator=model, param_grid=param_grid, verbose=2)
grid_result = grid.fit(X, Y)  
```

```python
# Results
print("Best: %.2f%% using %s" % (grid_result.best_score_*100, grid_result.best_params_))
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print("%.2f%% +/- %.2f%% with: %r" % (mean*100, stdev*100, param))  
```
